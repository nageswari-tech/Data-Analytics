```python
import random

def playgame():
    options = ["Rock", "Paper", "Scissors"]
    
    while True:
        try:
            # 1. Take user input
            user_input = input("\nEnter 1 for Rock, 2 for Paper, 3 for Scissors (or 'q' to quit): ")
            
            
            if user_input.lower() == 'q':
                print("Thanks for playing!")
                break
            
            
            choice_index = int(user_input) - 1
            
            
            if choice_index not in [0, 1, 2]:
                raise IndexError("Please choose a number between 1 and 3.")
            
            user_choice = options[choice_index]
            computer_choice = random.choice(options)
            
            print(f"You chose: {user_choice}")
            print(f"Computer chose: {computer_choice}")

            
            if user_choice == computer_choice:
                print("It's a tie!")
            elif (user_choice == "Rock" and computer_choice == "Scissors") or (user_choice == "Paper" and computer_choice == "Rock") or (user_choice == "Scissors" and computer_choice == "Paper"):
                print("You win!")
            else:
                print("You lose!")

        except ValueError:
            print("Invalid input! Please enter a valid number (1, 2, or 3).")
        except IndexError as e:
            print(e)

playgame()
```
